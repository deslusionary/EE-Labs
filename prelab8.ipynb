```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Variables
v_cc = 18
r_l = 4.7e3
r_e = 2.2e3

df_q78 = pd.read_csv("Q782023-11-14_204614.csv")
v_ce = np.linspace(0, 18, 100)
i_ce_ll =(v_cc - v_ce) / (r_l + r_e) # prelab 8 load line

plt.plot(df_q78["Keithley (V)"], df_q78["Keithley (A)"], '.')
plt.plot(v_ce, i_ce_ll, 'g-')
plt.title(r"$I_C$ vs. $V_{CE}$")
plt.xlabel(r"$V_{CE}$ (V)")
plt.ylabel(r"$I_C$ (A)")
plt.grid()
```
