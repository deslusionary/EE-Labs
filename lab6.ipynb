```python
df_lts_tf = pd.read_csv("pl6-transfer.txt")
df_lts_tf_filt = df_lts_tf[df_lts_tf["V(vgs)"] > -1.3]

plt.plot(df_part2["Vgs"], df_part2["Id"] * 1000, '.-', label="Experimental")
plt.plot(df_lts_tf_filt["V(vgs)"], df_lts_tf_filt["Id(J1)"] * 1e10, 'g-', label="Simulated")
plt.title(r"Transfer Characteristics")
plt.ylabel(r"$I_{DS}$ (mA)")
plt.xlabel(r"$V_{GS}$ (V)")
plt.legend()
plt.plot()
```

```python
# load line - theoretical FET
R = 1e3
df_drain_char = pd.read_csv("pl6-drain_char.txt", delimiter=r'\s+')

vds_ll = np.linspace(0, 16, num=100)
ids_ll = (10 - vds_ll) / R


plt.plot(df_drain_char["V(vds)"], df_drain_char["Id(J1)"], '.-', label=r"Drain Characteristics, $V_{GS}= -1$")
plt.plot(vds_ll, ids_ll, "g-", label=r"Load Line: $R = 1k \Omega$, $V_{DD}=10 V$")
plt.title(r"Theoretical Load Line Analysis: $V_{GS} = -1 V$")
plt.xlabel(r"$V_{DS}$ (V)")
plt.ylabel(r"$I_D$ (A)")
plt.ylim(0, df_drain_char["Id(J1)"].max() + 0.0002)
plt.legend()
print("max drain current is", df_drain_char["Id(J1)"].max() * 1e6)
```
